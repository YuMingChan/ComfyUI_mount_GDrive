```python
import os
import sys

# 檢查是否已經安裝好
def check_environment():
    try:
        import torch
        if torch.cuda.is_available():
            print("✅ 環境已準備就緒，CUDA 可用。")
            return True
    except:
        return False
    return False

if not check_environment():
    print("⚠️ 偵測到環境未就緒，執行強制修復...")
    # 這裡放你果兩行最強力嘅安裝指令
    !pip install -q torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu121
    # 執行一次重啟 kernel 嘅指令，強制刷新環境路徑
    os.kill(os.getpid(), 9)
```
